## Importações e definições iniciais

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")
```

## Criação das matrizes de transformação

```python
vectors = np.array([[2, -1], [0, 2], [3, 3]])


labels = ['v1 (2,-1)', 'v2 (0,2)', 'v3 (3,3)']

plt.figure(figsize=(5,5))
plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.gca().set_aspect('equal')

plt.xlim(-5, 5)
plt.ylim(-5, 5)

for vec, lab in zip(vectors, labels):
    plt.arrow(0,0, vec[0], vec[1], head_width=0.12, length_includes_head=True)
    plt.text(vec[0]*1, vec[1]*1, lab)

plt.title("Vetores Originais")
plt.grid(True)
plt.show()
```

# Aplicação de rotação



```python
Rot_90h = np.array([
    [0, -1],
    [1, 0]
])

vectors_rot = vectors @ Rot_90h.T

plt.figure(figsize=(6,6))
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.gca().set_aspect('equal')

plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.grid(True)

for vec, lab in zip(vectors, labels):
    plt.arrow(0,0, vec[0], vec[1], head_width=0.2, length_includes_head=True, color='blue', alpha=0.5)
    plt.text(vec[0]*1, vec[1]*1, lab, color='blue')

for vec, lab in zip(vectors_rot, labels):
    plt.arrow(0,0, vec[0], vec[1], head_width=0.2, length_includes_head=True, color='red')
    plt.text(vec[0]*1, vec[1]*1, lab + " (rot)", color='red')

plt.title("Vetores Rotacionados 90° Horário")
plt.show()

```

# Aplicação do Shear

```python
S = np.array([[1, 1],
              [0, 1]])

vectors_shear = vectors @ S.T

plt.figure(figsize=(6,6))
plt.axhline(0, color='black'); plt.axvline(0, color='black'); plt.gca().set_aspect('equal')
plt.xlim(-7, 7); plt.ylim(-7, 7)
plt.grid(True)

for vec, lab in zip(vectors, labels):
    plt.arrow(0, 0, vec[0], vec[1], head_width=0.12, length_includes_head=True)
    plt.text(vec[0]*1.05, vec[1]*1.05, lab)

for vec, lab in zip(vectors_shear, labels):
    plt.arrow(0, 0, vec[0], vec[1], head_width=0.12, length_includes_head=True, color='green')
    plt.text(vec[0]*1.05, vec[1]*1.05, lab + " (shear)")

plt.title("Originais (azul) e Shear (verde)")
plt.show()
```

# Vetores Rotacionados e Shearados

```python
vectors_RS = vectors_rot @ S.T

plt.figure(figsize=(6,6))
plt.axhline(0, color='black'); plt.axvline(0, color='black'); plt.gca().set_aspect('equal')
plt.xlim(-4, 4); plt.ylim(-4, 4)
plt.grid(True)

for vec, lab in zip(vectors, labels):
    plt.arrow(0, 0, vec[0], vec[1], head_width=0.12, length_includes_head=True)
    plt.text(vec[0]*1.05, vec[1]*1.05, lab)

for vec, lab in zip(vectors_RS, labels):
    plt.arrow(0, 0, vec[0], vec[1], head_width=0.12, 
              length_includes_head=True, color='purple')
    plt.text(vec[0]*1.05, vec[1]*1.05, lab + " (R→S)")

plt.title("Composição: Rotação → Shear")
plt.show()
```

## Vetores Shearados e depois Rotacionados

```python
vectors_SR = vectors_shear @ Rot_90h.T

plt.figure(figsize=(7,7))
plt.axhline(0, color='black'); plt.axvline(0, color='black'); plt.gca().set_aspect('equal')
plt.xlim(-7, 7); plt.ylim(-7, 7)
plt.grid(True)


for vec, lab in zip(vectors, labels):
    plt.arrow(0, 0, vec[0], vec[1], head_width=0.12, length_includes_head=True)
    plt.text(vec[0]*1.05, vec[1]*1.05, lab)


for vec, lab in zip(vectors_SR, labels):
    plt.arrow(0, 0, vec[0], vec[1], head_width=0.12, 
              length_includes_head=True, color='orange')
    plt.text(vec[0]*1.05, vec[1]*1.05, lab + " (S→R)")

plt.title("Composição: Shear → Rotação")
plt.show()
```

## Calculo dos determinantes

```python
det_R = np.linalg.det(Rot_90h)
det_S = np.linalg.det(S)

det_RS = np.linalg.det(Rot_90h @ S)   
det_SR = np.linalg.det(S @ Rot_90h)   

print("Determinante da Rotação (R):", det_R)
print("Determinante do Shear (S):  ", det_S)
print("Det R→S:", det_RS)
print("Det S→R:", det_SR)
```

## Vetor perpendicular e produto escalar

```python
x = np.array([3.0, 4.0])
R90 = np.array([[0, -1],
                [1,  0]])

y = R90 @ x 

dot = x @ y 

print("x =", x)
print("y =", y)
print("Produto escalar x·y =", dot)

# Plot
plt.figure(figsize=(6,6))
plt.axhline(0, color='black'); plt.axvline(0, color='black'); plt.gca().set_aspect('equal')
plt.xlim(-6, 6); plt.ylim(-6, 6)
plt.grid(True)

plt.arrow(0, 0, x[0], x[1], head_width=0.12, length_includes_head=True, color='blue')
plt.text(x[0]*1.05, x[1]*1.05, "x")

plt.arrow(0, 0, y[0], y[1], head_width=0.12, length_includes_head=True, color='red')
plt.text(y[0]*1.05, y[1]*1.05, "y (90°)")

plt.title("Vetor original x e vetor y rotacionado +90°")
plt.show()
```
